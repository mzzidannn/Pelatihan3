```python
#Menambah angka
a=np.array ([[1,8,8,2],[8,2,1,2],[10,1,4,7],[1,4,7,8],[8,2,3,4]])
print("array awal")
print(a)
a[1:4:2, 0:4] = a[1:4:2, 0:4]+1
print("baris ganjil +1")
print(a)
a[0:4, 0:4:2] = a[0:4, 0:4:2]+2
print("kolom genap  +2")
print(a)
```

```python
#Balik sumbu z
a=np.array([[[7,8,5],[4,7,7],[7,1,9],[5,4,10]],[[2,1,2],[9,8,5],[3,7,3],[4,9,9]],[[2,9,3],[2,4,8],[1,6,6],[8,1,3]]])

a=np.array_split(a,3)
a=np.concatenate((a[2],a[1],a[0]))
print(a)
```
